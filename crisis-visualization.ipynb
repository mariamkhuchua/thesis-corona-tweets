```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def visualize_crisis(df, crisis_type, indicators):
    """take country crisis df and save graphs"""
    plt.figure(figsize=(12,7))
    for indicator in indicators:
        plt.plot(df[indicator], "o-", label=indicator)
        plt.legend()
        plt.xticks(df["week"])   
        for x,y in enumerate(df[indicator]):
            label = str(y)
            plt.annotate(label,(x,y+0.5))   
    
    plt.tight_layout()
    plt.savefig(f"data/plots/{country}_{crisis_type}.jpg", pad_inches=1)
```

```python
countries = ["usa","uk"]
```

```python
for country in countries:
    econ = pd.read_csv(f"data/crisis-indicators/econ-{country}.csv", sep=";", decimal=",")
    med = pd.read_csv(f"data/crisis-indicators/med-{country}.csv", sep=";", decimal=",")
    econ_indicators = ["gdp change", "unemployment", "inflation"]
    med_indicators = ["new_cases","deaths"]
    visualize_crisis(econ,"econ",econ_indicators)
    visualize_crisis(med,"med",med_indicators)
```
